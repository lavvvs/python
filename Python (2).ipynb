```python
numbers = [1, 2, 2, 3, 4, 2, 3, 3, 3, 5]
frequency = {}
for num in numbers:
    if num in frequency:
        frequency[num] += 1
    else:
        frequency[num] = 1
most_frequent = ""
max_count = 0
for key, value in frequency.items():
    if value > max_count:
        max_count = value
        most_frequent = key
print("Most frequent element is:", most_frequent)
```

```python
nested= [1, [2, 3], [4, 5]]
def flat(lst):
    flat_list = []
    for i in lst:
        if isinstance(i, list): 
            flat_list.extend(flat(i))
        else:
            flat_list.append(i)
    return flat_list
flattened = flat(nested)
print(flattened)
```

```python
def is_palindrome(lst):
    if len(lst) <= 1:
        return True
    if lst[0] != lst[-1]:
        return False
    return is_palindrome(lst[1:-1])
lst = [1, 2, 3, 2, 1]
print("Palindrome" if is_palindrome(lst) else "Not a palindrome")
```

```python
list1 = [1, 3, 5, 7]
list2 = [2, 4, 6, 8]
merged_list = []
i = j = 0
while i < len(list1) and j < len(list2):
    if list1[i] < list2[j]:
        merged_list.append(list1[i])
        i += 1
    else:
        merged_list.append(list2[j])
        j += 1
while i < len(list1):
    merged_list.append(list1[i])
    i += 1

while j < len(list2):
    merged_list.append(list2[j])
    j += 1

print("Merged sorted list:", merged_list)
```

```python
scores = {'Alice': 85, 'Bob': 92, 'Charlie': 92, 'David': 78}
max_value = ""
max_keys = []
for key, value in scores.items():
    if (max_value =="") or (value > max_value):
        max_value = value
        max_keys = [key]
    elif value == max_value:
        max_keys.append(key)  
print("Maximum value:", max_value)
print("Key(s) with maximum value:", max_keys)
```

```python
words = ["apple", "banana", "apricot", "blueberry", "cherry", "avocado"]
grouped_words = {}
for word in words:
    first_letter = word[0]  
    if first_letter in grouped_words:
        grouped_words[first_letter].append(word)  
    else:
        grouped_words[first_letter] = [word]   
print(grouped_words)
```

```python
dict1 = {'a': 1, 'b': 2, 'c': 3}
dict2 = {'b': 2, 'a': 1, 'c': 4}
if len(dict1) != len(dict2):
    print("The dictionaries are not equal")
else:
    equal = True
    for key in dict1:
        if key not in dict2 or dict1[key] != dict2[key]:
            equal = False
            break
    print("The dictionaries are equal" if equal else "The dictionaries are not equal")
```
